```python
# Core libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import joblib  # for saving scaler

# Scikit-learn preprocessing tools
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# SMOTE - for fixing class imbalance
from imblearn.over_sampling import SMOTE

# Settings
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 5)
os.makedirs('../reports', exist_ok=True)
os.makedirs('../models', exist_ok=True)

print("Libraries imported successfully ✅")
```

```python
df = pd.read_csv('../data/raw/diabetes.csv')

print(f"Raw dataset shape : {df.shape}")
print(f"\nFirst 5 rows:")
df.head()
```

```python
# -------------------------------------------------------
# STEP 1: Replace impossible zero values with NaN
# -------------------------------------------------------
# These columns CANNOT be zero for a living person.
# Zeros here are actually missing/unrecorded values.
# We replace them with NaN so we can handle them properly.

zero_columns = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

df_clean = df.copy()  # Always work on a copy, never modify original!

for col in zero_columns:
    zero_count = (df_clean[col] == 0).sum()
    df_clean[col] = df_clean[col].replace(0, np.nan)
    print(f"  {col:25s}: {zero_count} zeros replaced with NaN")

print(f"\nMissing values after replacement:")
print(df_clean.isnull().sum())
```

```python
# -------------------------------------------------------
# STEP 2: Fill missing values with MEDIAN
# -------------------------------------------------------
# We use MEDIAN (not mean) because:
# - Mean is affected by outliers (e.g., Insulin has extreme values)
# - Median is the "middle value" — more robust for medical data
#
# We impute SEPARATELY for diabetic and non-diabetic patients
# because their medical values are naturally different!
# e.g., diabetic patients have higher glucose on average.
# Using a single median for everyone would be inaccurate.

print("Imputing missing values using GROUP MEDIAN (by Outcome):\n")

for col in zero_columns:
    # Calculate median for each group (0 and 1) separately
    median_by_group = df_clean.groupby('Outcome')[col].median()
    print(f"  {col} median → Non-Diabetic: {median_by_group[0]:.1f} | Diabetic: {median_by_group[1]:.1f}")

    # Fill NaN values with the median of their respective group
    df_clean[col] = df_clean.groupby('Outcome')[col].transform(
        lambda x: x.fillna(x.median())
    )

print(f"\nMissing values after imputation:")
print(df_clean.isnull().sum())
print("\n✅ No more missing values!")
```

```python
# -------------------------------------------------------
# Visualize: Before vs After imputation for key columns
# -------------------------------------------------------
# We want to make sure imputation didn't distort distributions

fig, axes = plt.subplots(2, 3, figsize=(16, 9))
axes = axes.flatten()

check_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for i, col in enumerate(check_cols):
    # Before imputation (original df with zeros)
    axes[i].hist(df[col], bins=30, alpha=0.5, color='#e74c3c',
                 label='Before (with zeros)', edgecolor='white')
    # After imputation
    axes[i].hist(df_clean[col], bins=30, alpha=0.5, color='#2ecc71',
                 label='After (imputed)', edgecolor='white')
    axes[i].set_title(f'{col}', fontweight='bold')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Count')
    axes[i].legend(fontsize=8)

# Hide the 6th unused subplot
axes[5].set_visible(False)

plt.suptitle('Before vs After Imputation', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('../reports/07_before_after_imputation.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Plot saved to reports/07_before_after_imputation.png")
```

```python
# -------------------------------------------------------
# STEP 3: Handle Outliers using IQR Capping
# -------------------------------------------------------
# IQR (Interquartile Range) = Q3 - Q1
# Values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR = outliers
#
# We use CAPPING (also called Winsorization):
# Instead of REMOVING outliers (losing data), we CAP them
# at the boundary values. This preserves data size.
#
# Why not remove? We only have 768 rows — we can't afford to lose data!

features = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
            'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

df_clean2 = df_clean.copy()

print("Outlier Capping using IQR method:\n")

for col in features:
    Q1 = df_clean2[col].quantile(0.25)   # 25th percentile
    Q3 = df_clean2[col].quantile(0.75)   # 75th percentile
    IQR = Q3 - Q1                         # Interquartile range

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers before capping
    outliers = ((df_clean2[col] < lower_bound) | (df_clean2[col] > upper_bound)).sum()

    # Cap values at boundaries
    df_clean2[col] = df_clean2[col].clip(lower=lower_bound, upper=upper_bound)

    print(f"  {col:28s}: {outliers:3d} outliers capped "
          f"[{lower_bound:.1f} — {upper_bound:.1f}]")

print("\n✅ Outlier capping complete!")
```

```python
# -------------------------------------------------------
# STEP 4: Feature Engineering
# -------------------------------------------------------
# Feature engineering = creating NEW meaningful features
# from existing ones to help the model learn better.
#
# We add:
# 1. BMI_Category    : Underweight/Normal/Overweight/Obese
# 2. Age_Group       : Young/Middle/Senior
# 3. Glucose_Level   : Normal/Prediabetes/Diabetes range
# 4. Insulin_Resistance: BMI × Glucose (interaction feature)

df_feat = df_clean2.copy()

# --- BMI Category (standard WHO classification) ---
def bmi_category(bmi):
    if bmi < 18.5:
        return 0   # Underweight
    elif bmi < 25:
        return 1   # Normal
    elif bmi < 30:
        return 2   # Overweight
    else:
        return 3   # Obese

df_feat['BMI_Category'] = df_feat['BMI'].apply(bmi_category)

# --- Age Group ---
def age_group(age):
    if age < 30:
        return 0   # Young
    elif age < 45:
        return 1   # Middle-aged
    else:
        return 2   # Senior

df_feat['Age_Group'] = df_feat['Age'].apply(age_group)

# --- Glucose Level Category ---
def glucose_level(glucose):
    if glucose < 100:
        return 0   # Normal
    elif glucose < 126:
        return 1   # Prediabetes
    else:
        return 2   # Diabetes range

df_feat['Glucose_Level'] = df_feat['Glucose'].apply(glucose_level)

# --- Insulin Resistance Score (interaction feature) ---
# High BMI + High Glucose = higher insulin resistance risk
df_feat['Insulin_Resistance'] = df_feat['BMI'] * df_feat['Glucose'] / 1000

print("New features added:")
print(f"  ✅ BMI_Category      : {df_feat['BMI_Category'].value_counts().to_dict()}")
print(f"  ✅ Age_Group         : {df_feat['Age_Group'].value_counts().to_dict()}")
print(f"  ✅ Glucose_Level     : {df_feat['Glucose_Level'].value_counts().to_dict()}")
print(f"  ✅ Insulin_Resistance: min={df_feat['Insulin_Resistance'].min():.2f}, "
      f"max={df_feat['Insulin_Resistance'].max():.2f}")

print(f"\nDataset shape after feature engineering: {df_feat.shape}")
df_feat.head()
```

```python
# -------------------------------------------------------
# STEP 5: Train-Test Split
# -------------------------------------------------------
# We split data into:
# - Training set (80%): model LEARNS from this
# - Test set    (20%): we EVALUATE model on this (unseen data)
#
# stratify=y ensures BOTH sets have the same class ratio
# random_state=42 ensures reproducibility (same split every time)

X = df_feat.drop('Outcome', axis=1)   # All columns except target
y = df_feat['Outcome']                 # Target column only

print(f"Features (X) shape : {X.shape}")
print(f"Target   (y) shape : {y.shape}")
print(f"Feature columns    : {list(X.columns)}")

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 20% for testing
    random_state=42,      # reproducibility
    stratify=y            # maintain class ratio in both splits
)

print(f"\nAfter Split:")
print(f"  X_train : {X_train.shape}  | y_train : {y_train.shape}")
print(f"  X_test  : {X_test.shape}   | y_test  : {y_test.shape}")
print(f"\nClass distribution in train set:")
print(y_train.value_counts())
print(f"\nClass distribution in test set:")
print(y_test.value_counts())
```

```python
# -------------------------------------------------------
# STEP 6: Fix Class Imbalance using SMOTE
# -------------------------------------------------------
# SMOTE = Synthetic Minority Oversampling TEchnique
#
# Instead of duplicating existing minority samples,
# SMOTE creates SYNTHETIC (artificial but realistic) new samples
# by interpolating between existing minority class samples.
#
# CRITICAL RULE: Apply SMOTE ONLY on training data!
# NEVER on test data — test data must reflect real world distribution.

print("Before SMOTE:")
print(f"  Training set class distribution:")
print(f"  {y_train.value_counts().to_dict()}")

smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print(f"\nAfter SMOTE:")
print(f"  Training set class distribution:")
unique, counts = np.unique(y_train_smote, return_counts=True)
print(f"  {dict(zip(unique, counts))}")
print(f"\n  X_train shape before SMOTE : {X_train.shape}")
print(f"  X_train shape after  SMOTE : {X_train_smote.shape}")
print(f"\n✅ Classes are now balanced!")
```

```python
# Visualize class balance before and after SMOTE
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Before SMOTE
before_counts = y_train.value_counts()
axes[0].bar(['Non-Diabetic', 'Diabetic'], before_counts.values,
            color=['#2ecc71', '#e74c3c'], edgecolor='black')
for i, v in enumerate(before_counts.values):
    axes[0].text(i, v + 3, str(v), ha='center', fontweight='bold')
axes[0].set_title('Before SMOTE (Training Set)', fontweight='bold')
axes[0].set_ylabel('Count')
axes[0].set_ylim(0, 500)

# After SMOTE
after_counts = pd.Series(y_train_smote).value_counts()
axes[1].bar(['Non-Diabetic', 'Diabetic'], after_counts.values,
            color=['#2ecc71', '#e74c3c'], edgecolor='black')
for i, v in enumerate(after_counts.values):
    axes[1].text(i, v + 3, str(v), ha='center', fontweight='bold')
axes[1].set_title('After SMOTE (Training Set)', fontweight='bold')
axes[1].set_ylabel('Count')
axes[1].set_ylim(0, 500)

plt.suptitle('Class Balance: Before vs After SMOTE', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../reports/08_smote_balance.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Plot saved to reports/08_smote_balance.png")
```

```python
# -------------------------------------------------------
# STEP 7: Feature Scaling using StandardScaler
# -------------------------------------------------------
# ML algorithms are sensitive to feature scales.
# Example: Age ranges 20-80, Insulin ranges 0-800
# Without scaling, Insulin would DOMINATE the model
# just because its numbers are bigger — that's wrong!
#
# StandardScaler transforms each feature to:
# Mean = 0, Standard Deviation = 1
# Formula: z = (x - mean) / std
#
# CRITICAL RULE: Fit scaler on TRAINING data ONLY,
# then transform both train and test with same scaler.
# Never fit on test data — that would cause "data leakage"!

scaler = StandardScaler()

# Fit on training data and transform it
X_train_scaled = scaler.fit_transform(X_train_smote)

# Transform test data using the SAME scaler (don't fit again!)
X_test_scaled = scaler.transform(X_test)

print("Feature Scaling Applied ✅")
print(f"\nBefore scaling (first row of X_train):")
print(np.round(X_train_smote.iloc[0].values, 3))
print(f"\nAfter scaling (same row):")
print(np.round(X_train_scaled[0], 3))
print(f"\nScaled data stats (should be ~mean=0, std=1):")
print(f"  Mean : {X_train_scaled.mean():.6f}")
print(f"  Std  : {X_train_scaled.std():.6f}")

# Save the scaler — we'll need it in Phase 8 for predictions!
joblib.dump(scaler, '../models/scaler.pkl')
print(f"\n✅ Scaler saved to models/scaler.pkl")
```

```python
# -------------------------------------------------------
# Save everything we need for Phase 5 (Model Training)
# -------------------------------------------------------
import os

os.makedirs('../data/processed', exist_ok=True)

# Save as numpy arrays (efficient for ML)
np.save('../data/processed/X_train.npy', X_train_scaled)
np.save('../data/processed/X_test.npy', X_test_scaled)
np.save('../data/processed/y_train.npy', y_train_smote)
np.save('../data/processed/y_test.npy', y_test.values)

# Also save cleaned dataframe as CSV for reference
df_feat.to_csv('../data/processed/diabetes_cleaned.csv', index=False)

# Save feature names for later use
feature_names = list(X.columns)
joblib.dump(feature_names, '../models/feature_names.pkl')

print("All processed data saved ✅")
print(f"\nSaved files:")
print(f"  📁 data/processed/X_train.npy       → {X_train_scaled.shape}")
print(f"  📁 data/processed/X_test.npy        → {X_test_scaled.shape}")
print(f"  📁 data/processed/y_train.npy       → {y_train_smote.shape}")
print(f"  📁 data/processed/y_test.npy        → {y_test.shape}")
print(f"  📁 data/processed/diabetes_cleaned.csv")
print(f"  📁 models/scaler.pkl")
print(f"  📁 models/feature_names.pkl")
```
